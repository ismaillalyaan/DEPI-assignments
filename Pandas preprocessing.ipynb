## worldometer_data dataset

```python
df6 = pd.read_csv('worldometer_data.csv')
df6.head()
```

```python
df6.info()
```

```python
df6.isna().sum()
```

Continent is string column so we can replace the missing value with the most repeated one
But for other columns we have numnerical values so we can just replace the missing values with the mean of the columns

```python
df6['Continent'] = df6['Continent'].fillna(df6['Continent'].mode()[0])
df6['WHO Region'] = df6['WHO Region'].fillna(df6['WHO Region'].mode()[0])
nums = df6.select_dtypes(include=['float64', 'int64']).columns
df6[nums] = df6[nums].fillna(df6[nums].mean())
```

```python
df6.isna().sum()
```

```python
df6.drop_duplicates(inplace = True)
```
